## Economics Data EDA

```
import pandas as pd
import plotly.graph_objs as go
from econ_eda import plot_series_by_state
```

```
econ_state = pd.read_csv("/work/Datasets/econ_state.csv")
econ_state = econ_state.set_index("state")
state_quarter_gdp = econ_state.iloc[:, :47].T
state_quarter_capita = econ_state.iloc[:, 47: 314].T
state_month_ur = econ_state.iloc[:, 314:].T

```

```
gdp_quarter_series = plot_series_by_state(state_quarter_gdp, title_text = "US South Region Quarterly GDP Series")
capita_quarter_series = plot_series_by_state(state_quarter_capita, title_text = "US South Region Quarterly Income per Capita Series")
ur_month_series = plot_series_by_state(state_month_ur, title_text = "US South Region Monthly Unemployment Rate Series")
```

```
gdp_quarter_series.show()
```

```
capita_quarter_series.show()
```

```
ur_month_series.show()
```
